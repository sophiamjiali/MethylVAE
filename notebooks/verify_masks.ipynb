```python
import os
import numpy as np
from PIL import Image

dir_path = "/Users/sophiali/Desktop/MethylVAE/patches"
```

```python
def find_masks_with_ones(directory_path):
    masks_with_ones = []
    
    # Iterate through all files in the directory
    for filename in os.listdir(directory_path):
        if filename.endswith(".png"):
            file_path = os.path.join(directory_path, filename)
            
            # Open the image and convert to numpy array
            with Image.open(file_path) as img:
                mask_data = np.array(img)
                
                # Check if the value 1 exists anywhere in the array
                if np.any(mask_data == 1):
                    masks_with_ones.append(filename)
                    
    return masks_with_ones
```

```python
result = find_masks_with_ones(dir_path)

print(f"Found {len(result)} masks containing the value 1:")
for name in result:
    print(name)
```
